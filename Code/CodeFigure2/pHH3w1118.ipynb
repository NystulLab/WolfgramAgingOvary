Plotting the number of pHH3+, Fas3+ Cells in 1 vs 6 week w1118(x)

Note: You will need to run the StatspHH3w1118.Rmd file first, or annotate out the stats code

```python
csv_path = "../../Data/DataFigure2/w1118pHH3_n.csv"
df = pd.read_csv(csv_path)
```

```python
df['Weeks'] = df['Weeks'].map({1: '1 Week', 6: '6 Weeks'})
```

```python
df.head(7)
```

```python
stats_path = "../../Results/ResultsFigure2/StatspHH3Poisson.csv"
stats = pd.read_csv(stats_path)

stats["WeekNum"] = stats["comparison"].str.extract(r"(\d+)").astype(int)
star = stats.loc[0, "stars"]
week_to_annotate = stats.loc[0, "WeekNum"]


df["WeekNum"] = df["Weeks"].str.extract(r"(\d+)").astype(int)

week_order = sorted(df["WeekNum"].unique())
label_order = (
    df.drop_duplicates("WeekNum")
      .sort_values("WeekNum")["Weeks"]
      .tolist()
)

fig, ax = plt.subplots(figsize=(3, 5))
sns.set(style="white")

sns.barplot(
    data=df,
    x="Weeks",
    y="pHH3",
    order=label_order,
    palette="gray",
    capsize=0.1,
    err_kws={"color": "black", "linewidth": 3},
    edgecolor="black",
    linewidth=3,
    width=0.7,
    ax=ax
)

sns.stripplot(
    data=df,
    x="Weeks",
    y="pHH3",
    order=label_order,
    color="black",
    alpha=0.5,
    size=6,
    jitter=True,
    ax=ax
)

y_max = df.groupby("WeekNum")["pHH3"].max()
offset = 0.1

x_position = week_order.index(week_to_annotate)

ax.text(
    x=x_position,
    y=y_max.loc[week_to_annotate] + offset,
    s=star,
    ha="center",
    va="bottom",
    fontsize=25,
    fontweight="bold"
)


plt.ylim(0, 1.4)
plt.ylabel("pHH3+, Fas3+ cells", fontsize=15, weight="bold")
plt.xlabel("")

ax.tick_params(width=3, labelsize=12)
ax.spines["left"].set_linewidth(3)
ax.spines["bottom"].set_linewidth(3)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

results_directory = '../../Results'
filename = os.path.join(results_directory, 'ResultsFigure2', 'pHH3w1118.pdf')
os.makedirs(os.path.dirname(filename), exist_ok=True)
plt.savefig(
    filename,
    format='pdf',
    bbox_inches='tight'
)

plt.tight_layout()
plt.show()
```
